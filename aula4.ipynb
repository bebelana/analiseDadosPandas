```python
# Importando biblioteca pandas do python
import pandas as pd
```

```python
# Carregando um conjunto de dados
df = pd.read_excel("/ca-2021-02.xlsx")
```

```python
# Inserindo coluna
df['Ativo'] = True
df.head(5)
```

```python
df['Obs'] = ["Melhor Estado" if municipio == 'MG' else '' for municipio in df['Estado - Sigla']]
filtrado = df.loc[df['Estado - Sigla'] == 'MG']
display(filtrado)
 
```

```python
import numpy as np
```

```python
#por Leandro e alterado por mim adicionando outra verificação
# Adicionando a nova coluna "Status preço"
df['Status preco'] = np.where((df['Valor de Venda'] > 5.5) & (df['Produto'] == 'GASOLINA'), 'Caro', 'Barato')
```

```python
display(df[['Produto', 'Obs', 'Valor de Venda', 'Status preco']])
```

```python
# Criando Data Frame 
habitantes_cidade_df = pd.read_csv("/ibge_num_habitantes_estimado.csv", sep=';' )
display(habitantes_cidade_df)
```

```python
# Deixar colunas com  nomes iguais 
habitantes_cidade_df.rename(columns={'Estado':'Estado - Sigla'}, inplace=True)
display(habitantes_cidade_df)
```

```python
# Unindo os dois dataframes pelo atributo "Municipio"
df = pd.merge(df, habitantes_cidade_df, on=['Municipio', 'Estado - Sigla'])
display(df)
```

```python
# Criando uma nova coluna com o número de postos 
num_postos = df['Revenda'].count()
df['Numero de Postos'] = num_postos
display(df)
```

```python
# calcular postos por habitantes
df['Postos por habitantes'] = df['Numero de Postos'] / df['NumHabitantes2021']
display(df)
```

```python
df.dropna(axis='columns', inplace=True)
print(df.info())
```

```python
colunas=['Regiao - Sigla', 'Nome da Rua', 'Numero Rua', 
         'Bairro', 'Cep', 'Produto', 'Data da Coleta', 'Valor de Venda',
         'Unidade de Medida', 'Bandeira', 'Ativo', 'Status preco', 'Obs','NumHabitantes2021_x','NumHabitantes2021_y' ]
df.drop(labels=colunas, axis=1, inplace=True)
print(df.info())
```
