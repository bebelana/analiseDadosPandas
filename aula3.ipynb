```python
# Importando biblioteca pandas do python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregando um conjunto de dados
df = pd.read_excel("/ca-2021-02.xlsx")
```

```python
# Utilizando colunas especificas 
columns_df = df[['Revenda', 'Municipio','Produto', 'Valor de Venda', 'Bairro']]
display(columns_df)
```

```python
# Filtrando por uma condicao, buscando por postos de BH com gasolina
gasolinaBH_df = columns_df.loc[(columns_df['Municipio'] == 'BELO HORIZONTE') & (columns_df['Produto'] == 'GASOLINA')]
display(gasolinaBH_df)
```

```python
#nova informação a partir de uma coluna existente
gasolinaBH_df['Imposto'] = gasolinaBH_df['Valor de Venda'] * 0.10
display(gasolinaBH_df)
```

```python
# Ordernar o data frame pela coluna valor de venda
display(gasolinaBH_df.sort_values(by='Valor de Venda', ascending=True))
```

```python
gasolinaBH_df.loc[(gasolinaBH_df['Bairro'] == 'PRADO')]
```

```python
# Media da Gasolina e Gasolina Aditivada de BH no bairro Prado
df.loc[((df['Produto'] == 'GASOLINA') | (df['Produto'] == 'GASOLINA ADITIVADA'))  &
       (df['Bairro'] == 'PRADO') &
       (df['Municipio']=='BELO HORIZONTE'), ['Valor de Venda']].mean().round(2)
```

```python
# como mostrar media de valor de venda por combustivel
media_produto = columns_df[['Produto', 'Valor de Venda']].groupby('Produto').mean()
display(media_produto.style.format("{:.2f}"))
```

```python
# Quero adicionar uma coluna de valor booleano no combustiveis_df 
# Chamada "Ativo" que, por padrão, vai ser True para 
# TODAS as linhas
df['Ativo'] = True
print(df.info())
display(df.head())
```

```python
df['Obs'] = ["Melhor Estado" if municipio == 'MG' else '' for municipio in df['Estado - Sigla']]
```

```python
# Exportar para excel
gasolinaBH_df.to_excel("/gasolinaBH_df.xlsx")
```

```python
# Criando o gráfico de dispersão dos valores de compra e venda
df.plot(x='Valor de Compra', y='Valor de Venda', kind='scatter')
plt.show()
```

```python
# Calculando a média do valor de venda por produto
media_produto = df[['Produto', 'Valor de Venda']].groupby('Produto').mean().round(2)
print(media_produto)
```

```python
# Calculando a média do valor de venda por município
media_municipio = df[['Municipio', 'Valor de Venda']].groupby('Municipio').mean().round(2)
print(media_municipio)
```

```python
# Calculando a média do valor de venda por estado
media_estado = df[['Estado - Sigla', 'Valor de Venda']].groupby('Estado - Sigla').mean().round(2)
print(media_estado)
```

```python
# Criando o gráfico de barras do valor de venda por estado
media_estado.plot(kind='bar', legend=None)
plt.ylabel('Valor de Venda Médio')
plt.show()
```
